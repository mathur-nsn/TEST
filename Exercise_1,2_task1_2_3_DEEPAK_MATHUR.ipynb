```python
# Join orders_df and returns_df on 'Order ID'
# Using a left join to keep all orders and add return information where available
orders_returns_df = pd.merge(orders_df, returns_df[['Order ID', 'Returned']], on='Order ID', how='right')

# Join the result with people_df on 'Region' using a right join
# This will keep all regions from people_df and add matching order/return information
final_dataset_right_join = pd.merge(orders_returns_df, people_df, on='Region', how='right')

print("Final dataset created by joining orders_df, returns_df, and people_df using a right join:")
display(final_dataset_right_join.head())
print(f"\nShape of the final dataset with right join: {final_dataset_right_join.shape}")
```

**Inner Join**

```python
# Join orders_df and returns_df on 'Order ID' using an inner join
# This will keep only orders that have a matching return record
orders_returns_inner_df = pd.merge(orders_df, returns_df[['Order ID', 'Returned']], on='Order ID', how='inner')

# Join the result with people_df on 'Region' using an inner join
# This will keep only the rows where there is a match in both dataframes on Region
final_dataset_inner_join = pd.merge(orders_returns_inner_df, people_df, on='Region', how='inner')

print("Final dataset created by joining orders_df, returns_df, and people_df using inner joins:")
display(final_dataset_inner_join.head())
print(f"\nShape of the final dataset with inner join: {final_dataset_inner_join.shape}")
```

**Outer Join**

```python
# Join orders_df and returns_df on 'Order ID' using an outer join
# This will keep all rows from both dataframes
orders_returns_outer_df = pd.merge(orders_df, returns_df[['Order ID', 'Returned']], on='Order ID', how='outer')

# Join the result with people_df on 'Region' using an outer join
# This will keep all rows from both dataframes on Region
final_dataset_outer_join = pd.merge(orders_returns_outer_df, people_df, on='Region', how='outer')

print("Final dataset created by joining orders_df, returns_df, and people_df using outer joins:")
display(final_dataset_outer_join.head())
print(f"\nShape of the final dataset with outer join: {final_dataset_outer_join.shape}")
```
